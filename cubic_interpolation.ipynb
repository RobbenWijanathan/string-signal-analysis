```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
```

Single-Instrument Notes

```python
# Bass A4 (A2)

data = pd.read_csv("./plot_data/Bass A4 (A2).csv", header=None) 
x = data.iloc[:, 0].values
y = data.iloc[:, 1].values

cs = CubicSpline(x, y)

x_smooth = np.linspace(min(x), max(x), 500)
y_smooth = cs(x_smooth)

plt.figure(figsize=(5, 10))
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_smooth, y_smooth, color='blue', linewidth=2, label='Cubic Spline')
plt.title("Bass A4 (A2) Cubic Interpolation", fontsize=16, weight='bold')
plt.xlabel("Time [s]", fontsize=14)
plt.ylabel("Amplitude", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tick_params(axis='both', labelsize=12)
plt.gcf().patch.set_facecolor('#f9f9f9')
plt.tight_layout()
plt.show()

print(cs.c)
```

```python
# Violin A4-E5

data = pd.read_csv("./plot_data/Violin A4-E5.csv", header=None)
x = data.iloc[:, 0].values
y = data.iloc[:, 1].values

sorted_indices = np.argsort(x)
x = x[sorted_indices]
y = y[sorted_indices]

cs = CubicSpline(x, y)

x_smooth = np.linspace(min(x), max(x), 500)
y_smooth = cs(x_smooth)

plt.figure(figsize=(8, 10))
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_smooth, y_smooth, color='blue', linewidth=2, label='Cubic Spline')
plt.title("Violin A4-E5 Cubic Interpolation", fontsize=16, weight='bold')
plt.xlabel("Time [s]", fontsize=14)
plt.ylabel("Amplitude", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tick_params(axis='both', labelsize=12)
plt.gcf().patch.set_facecolor('#f9f9f9')
plt.tight_layout()
plt.legend()
plt.show()

print(cs.c)
```
