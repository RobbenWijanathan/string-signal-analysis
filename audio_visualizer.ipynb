```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

Single-Instrument Notes

```python
# Bass A4 (A2)

sample_rate, data = wavfile.read("./audio/Bass A4 (A2).wav")

if data.ndim > 1:
    data = data[:, 0]  # use left channel
    
time = np.arange(len(data)) / sample_rate

start_time = 0.45
end_time = 0.5

start_index = int(start_time * sample_rate)
end_index = int(end_time * sample_rate)

segment = data[start_index:end_index]
segment_time = time[start_index:end_index]

plt.figure(figsize=(10, 6))
plt.plot(segment_time, segment, color=plt.cm.cool(0.5), linewidth=1.5)
plt.title("Bass A4 (A2) (0.45 - 0.5 seconds)", fontsize=16, weight='bold')
plt.xlabel("Time [s]", fontsize=14)
plt.ylabel("Amplitude", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tick_params(axis='both', labelsize=12)
plt.locator_params(axis='x', nbins=20)
plt.locator_params(axis='y', nbins=20)
plt.gcf().patch.set_facecolor('#f9f9f9')
plt.tight_layout()
plt.show()
```

Multiple-Instrument Chords

```python
# Violin A4-E5 (Perfect Fifth) + Guitar A Major

sample_rate, data = wavfile.read("./audio/Violin A4-E5 (Perfect Fifth) + Guitar A Major.wav")

if data.ndim > 1:
    data = data[:, 0]  # use left channel
    
time = np.arange(len(data)) / sample_rate

start_time = 0.45
end_time = 0.5

start_index = int(start_time * sample_rate)
end_index = int(end_time * sample_rate)

segment = data[start_index:end_index]
segment_time = time[start_index:end_index]

plt.figure(figsize=(10, 6))
for i in range(len(segment_time)-1):
    plt.plot(segment_time[i:i+2], segment[i:i+2],
             color=plt.cm.inferno(i / len(segment_time) - 0.3), linewidth=1.5)
plt.title("Violin A4-E5 + Guitar A Major (0.45 - 0.5 seconds)", fontsize=16, weight='bold')
plt.xlabel("Time [s]", fontsize=14)
plt.ylabel("Amplitude", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tick_params(axis='both', labelsize=12)
plt.locator_params(axis='x', nbins=20)
plt.locator_params(axis='y', nbins=20)
plt.gcf().patch.set_facecolor('#f9f9f9')
plt.tight_layout()
plt.show()
```
